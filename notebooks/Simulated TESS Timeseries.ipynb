```python
from tesszap.tess import StampCadenceTimeseries, FFICadenceTimeseries
from tesszap.strategies import Central
from tesszap.models import Transit
%matplotlib inline
```

Let's simulate a general timeseries, that's been stacked together using the a Central 8/10 strategy:

```python
# let's create a stacking strategy object
strategy = Central(10)

# and let's make a model object
transit = Transit(a_over_rs=5, period=1.6)

# now, let's check out a particular TESS magnitude
tmag = 12
```

```python
# let's create a Stamp timeseries
stamp = StampCadenceTimeseries(model=transit, tmin=-27.6/2, tmax=27.6/2, tmag=tmag)
stamp.stack(strategy)
stamp.plot(xlim=[-0.5, 0.5])
```

```python
# let's create an FFI timeseries and check it out
ffi = FFICadenceTimeseries(model=transit, tmin=-27.6/2, tmax=27.6/2, tmag=tmag)
ffi.stack(strategy)
ffi.plot(xlim=[-0.5, 0.5])
```

Just to inspect everything, let's look at a movie sliding through the light curves.

```python
stamp.movie(filename='stamp.mp4')
ffi.movie(filename='ffi.mp4')
```
